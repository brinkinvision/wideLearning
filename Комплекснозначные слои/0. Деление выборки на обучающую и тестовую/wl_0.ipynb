```python
import pandas as pd
```

```python
f = pd.read_csv('wine1.csv')
pd.set_option('display.max_rows', None)
pd.set_option('max_seq_items', None)
```

```python
f = f.reset_index() 
f
```

```python
max_v = []
for i in f:
    max_v.append(f[f'{i}'].max())
    
del max_v[0] 
```

```python
del max_v[0] 
max_v
```

```python
min_v = []
for i in f:
    min_v.append(f[f'{i}'].min())
    
del min_v[0] 
```

```python
del min_v[0] 
min_v
```

```python
for i in range(len(f['Alcohol'])):
    f['Alcohol'][i] = 2*((f['Alcohol'][i] - min_v[0])/(max_v[0] - min_v[0]))-1
```

```python
for i in range(len(f['Malic.acid'])):
    f['Malic.acid'][i] = 2*((f['Malic.acid'][i] - min_v[1])/(max_v[1] - min_v[1]))-1
```

```python
for i in range(len(f['Ash'])):
    f['Ash'][i] = 2*((f['Ash'][i] - min_v[2])/(max_v[2] - min_v[2]))-1
```

```python
for i in range(len(f['Acl'])):
    f['Acl'][i] = 2*((f['Acl'][i] - min_v[3])/(max_v[3] - min_v[3]))-1
```

```python
f['Mg'] = f['Mg'].astype('float')
for i in range(len(f['Mg'])):
    f['Mg'][i] = 2*((f['Mg'][i] - min_v[4])/(max_v[4] - min_v[4]))-1
```

```python
for i in range(len(f['Phenols'])):
    f['Phenols'][i] = 2*((f['Phenols'][i] - min_v[5])/(max_v[5] - min_v[5]))-1
```

```python
for i in range(len(f['Flavanoids'])):
    f['Flavanoids'][i] = 2*((f['Flavanoids'][i] - min_v[6])/(max_v[6] - min_v[6]))-1
```

```python
for i in range(len(f['Nonflavanoid.phenols'])):
    f['Nonflavanoid.phenols'][i] = 2*((f['Nonflavanoid.phenols'][i] - min_v[7])/(max_v[7] - min_v[7]))-1
```

```python
for i in range(len(f['Proanth'])):
    f['Proanth'][i] = 2*((f['Proanth'][i] - min_v[8])/(max_v[8] - min_v[8]))-1
```

```python
for i in range(len(f['Color.int'])):
    f['Color.int'][i] = 2*((f['Color.int'][i] - min_v[9])/(max_v[9] - min_v[9]))-1
```

```python
for i in range(len(f['Hue'])):
    f['Hue'][i] = 2*((f['Hue'][i] - min_v[10])/(max_v[10] - min_v[10]))-1
```

```python
for i in range(len(f['OD'])):
    f['OD'][i] = 2*((f['OD'][i] - min_v[11])/(max_v[11] - min_v[11]))-1
```

```python
f['Proline'] = f['Proline'].astype('float')
for i in range(len(f['Proline'])):
    f['Proline'][i] = 2*((f['Proline'][i] - min_v[12])/(max_v[12] - min_v[12]))-1
```
